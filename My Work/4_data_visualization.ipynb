```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Load the dataset
file_path = '/content/cleaned_dataset.csv'
data = pd.read_csv(file_path)

data.head()
```

# a. Visualize Categorical Variables:

Plot bar charts or pie charts to visualize the distribution of categorical variables such as smoking status (Smoker), physical activity (PhysActivity), fruit consumption (Fruits), and vegetable consumption (Veggies).

```python
import matplotlib.pyplot as plt

# Define a function to plot both bar and pie charts for categorical variables
def plot_categorical_distribution(data, column_name, title):
    # Count the occurrences of each category
    value_counts = data[column_name].value_counts()

    # Create a figure with two subplots: Bar and Pie
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Bar chart
    value_counts.plot(kind='bar', ax=axes[0], color='skyblue')
    axes[0].set_title(f'{title} - Bar Chart')
    axes[0].set_xlabel(column_name)
    axes[0].set_ylabel('Count')

    # Pie chart
    value_counts.plot(kind='pie', ax=axes[1], autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'], startangle=90)
    axes[1].set_title(f'{title} - Pie Chart')
    axes[1].set_ylabel('')  # Remove y-label for pie chart

    plt.tight_layout()
    plt.show()

# Plot distributions for the requested categorical variables
plot_categorical_distribution(data, 'Smoker', 'Smoking Status')
plot_categorical_distribution(data, 'PhysActivity', 'Physical Activity')
plot_categorical_distribution(data, 'Fruits', 'Fruit Consumption')
plot_categorical_distribution(data, 'Veggies', 'Vegetable Consumption')
```


# b. Compare Heart Disease Across Categories:

Use grouped bar charts or stacked bar charts to compare the prevalence of heart disease across different categorical variables.

```python
# Define a function to plot grouped bar charts comparing heart disease prevalence across categorical variables
def plot_heart_disease_comparison(data, category, title):
    # Create a crosstab to count occurrences of heart disease (1) and no heart disease (0) within each category
    heart_disease_crosstab = pd.crosstab(data[category], data['HeartDiseaseorAttack'])

    # Plot grouped bar chart
    heart_disease_crosstab.plot(kind='bar', stacked=False, color=['lightblue', 'salmon'], figsize=(10, 6))
    plt.title(f'Heart Disease Prevalence vs {title}')
    plt.xlabel(title)
    plt.ylabel('Count')
    plt.legend(['No Heart Disease', 'Heart Disease'])
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Compare heart disease prevalence across the requested categorical variables
plot_heart_disease_comparison(data, 'Smoker', 'Smoking Status')
plot_heart_disease_comparison(data, 'PhysActivity', 'Physical Activity')
plot_heart_disease_comparison(data, 'Fruits', 'Fruit Consumption')
plot_heart_disease_comparison(data, 'Veggies', 'Vegetable Consumption')
```

The grouped bar charts compare the prevalence of heart disease across various categories:

Heart Disease vs Smoking Status:
Smokers have a higher prevalence of heart disease compared to non-smokers.

Heart Disease vs Physical Activity:
Individuals who are physically inactive show a higher prevalence of heart disease compared to those who are physically active.

Heart Disease vs Fruit Consumption:
There seems to be a slight difference in heart disease prevalence between those who consume fruits and those who do not.

Heart Disease vs Vegetable Consumption:
Similar to fruit consumption, there is a slight variation in heart disease prevalence between those who consume vegetables and those who don't.

These visualizations help highlight the associations between lifestyle factors and heart disease.

# Explore other Visualization Techniques

a. Display Distributions of Continuous Variables:
Plot histograms or box plots to visualize the distributions of continuous variables such as BMI, mental health days (Menthlth), and physical health days (PhysHlth).

```python
# Define a function to plot histograms and box plots for continuous variables
def plot_continuous_distribution(data, column_name, title):
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Histogram
    data[column_name].plot(kind='hist', bins=20, ax=axes[0], color='skyblue', edgecolor='black')
    axes[0].set_title(f'{title} - Histogram')
    axes[0].set_xlabel(column_name)
    axes[0].set_ylabel('Frequency')

    # Box plot
    data[column_name].plot(kind='box', ax=axes[1], vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
    axes[1].set_title(f'{title} - Box Plot')
    axes[1].set_xlabel(column_name)

    plt.tight_layout()
    plt.show()

# Plot distributions for BMI, MentHlth, and PhysHlth
plot_continuous_distribution(data, 'BMI', 'BMI (Body Mass Index)')
plot_continuous_distribution(data, 'MentHlth', 'Mental Health Days')
plot_continuous_distribution(data, 'PhysHlth', 'Physical Health Days')
```

# b. Heatmaps and Correlation Plots: Explore Variable Relationships
Use heatmaps or correlation matrices to visualize relationships and correlations between variables.

```python
#'BMI', 'MentHlth', and 'PhysHlth' are continuous variables
continuous_vars = ['BMI', 'MentHlth', 'PhysHlth']

# Calculate the correlation matrix
correlation_matrix = data[continuous_vars].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Continuous Variables')
plt.show()
```

# c. Scatter Plots: Investigate Relationships Between Variables
Plot scatter plots to explore relationships between pairs of variables, focusing on those potentially associated with heart disease.

```python
# List of continuous variables
continuous_vars = ['BMI', 'Age', 'MentHlth', 'PhysHlth', 'GenHlth']

# Scatter plots for continuous variables
plt.figure(figsize=(15, 12))
for i in range(len(continuous_vars)):
    for j in range(i + 1, len(continuous_vars)):
        plt.subplot(len(continuous_vars) - 1, len(continuous_vars) - 1, i * (len(continuous_vars) - 1) + j)
        # Now 'df' is defined and can be used here
        sns.scatterplot(x=data[continuous_vars[i]], y=data[continuous_vars[j]], hue=data['HeartDiseaseorAttack'], palette='viridis', alpha=0.7)
        plt.title(f'{continuous_vars[i]} vs {continuous_vars[j]}')
        plt.xlabel(continuous_vars[i])
        plt.ylabel(continuous_vars[j])
        plt.legend(title='Heart Disease', loc='best')

plt.tight_layout()
plt.show()
```

# e. Violin Plots: Combine aspects of box plots and density plots to show the distribution of data.

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Smoker', y='BMI', data=data, palette='coolwarm')
plt.title('Distribution of BMI by Smoking Status')
plt.xlabel('Smoker (0 = No, 1 = Yes)')
plt.ylabel('BMI')
plt.show()
```

```python
# Create violin plots for each continuous variable
plt.figure(figsize=(12, 10))
for i, var in enumerate(continuous_vars):
    plt.subplot(len(continuous_vars), 1, i + 1)
    sns.violinplot(x='HeartDiseaseorAttack', y=var, data=data, palette='viridis')
    plt.title(f'Distribution of {var} by Heart Disease')
    plt.xlabel('Heart Disease')
    plt.ylabel(var)

plt.tight_layout()
plt.show()
```

```python
# Subset the dataset to include only the relevant variables
variables = ['Age','BMI', 'MentHlth', 'PhysHlth', 'HeartDiseaseorAttack']
subset_data = data[variables]

sns.pairplot(subset_data, hue='HeartDiseaseorAttack', palette='coolwarm', diag_kind='kde', markers=["o", "s"])
plt.suptitle('Pairwise Relationships Between Variables', y=1.02)  # Title with proper spacing
plt.show()
```
