```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/cleaned_dataset.csv')
```

```python
df.head()
```

# 1. **Univariate analysis**

a. Distributions of Individual Variables:

```python

# List of variables to visualize
variables_to_plot = ['Age', 'BMI', 'MentHlth', 'PhysHlth']

# Plotting histograms for the selected variables
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.suptitle('Distribution of Key Variables')

# Histogram for Age
axes[0, 0].hist(df['Age'], bins=10, color='skyblue')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Frequency')

# Histogram for BMI
axes[0, 1].hist(df['BMI'], bins=10, color='lightgreen')
axes[0, 1].set_title('BMI Distribution')
axes[0, 1].set_xlabel('BMI')
axes[0, 1].set_ylabel('Frequency')

# Histogram for MentHlth
axes[1, 0].hist(df['MentHlth'], bins=10, color='salmon')
axes[1, 0].set_title('Mental Health Days Distribution')
axes[1, 0].set_xlabel('Mental Health Days')
axes[1, 0].set_ylabel('Frequency')

# Histogram for PhysHlth
axes[1, 1].hist(df['PhysHlth'], bins=10, color='lightcoral')
axes[1, 1].set_title('Physical Health Days Distribution')
axes[1, 1].set_xlabel('Physical Health Days')
axes[1, 1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Display the plots
plt.show()
```

b. Prevalence of Health Conditions:

```python
# List of health conditions to visualize
health_conditions = ['HighBP', 'HighChol', 'HeartDiseaseorAttack', 'Stroke', 'Diabetes']

# Create subplots for bar charts
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
fig.suptitle('Prevalence of Health Conditions')

# Loop through the health conditions to create bar charts
for i, condition in enumerate(health_conditions):
    row = i // 3
    col = i % 3

    condition_counts = df[condition].value_counts()

    # Plot bar chart for each condition
    axes[row, col].bar(condition_counts.index, condition_counts.values, color='lightblue')
    axes[row, col].set_title(f'{condition} Prevalence')
    axes[row, col].set_xlabel('Condition Present (1 = Yes, 0 = No)')
    axes[row, col].set_ylabel('Count')

# Adjust layout for better visibility
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Display the bar charts
plt.show()
```

```python
print("PLOTTING PIE CHART TO SHOW PRELEVANCE")
# Plot pie chart for HighBP
plt.figure(figsize=(8, 6))
df['HighBP'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Prevalence of High Blood Pressure')
plt.axis('equal')
plt.show()

# Plot pie chart for HighChol
plt.figure(figsize=(8, 6))
df['HighChol'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Prevalence of High Cholesterol')
plt.axis('equal')
plt.show()

# Plot pie chart for Stroke
plt.figure(figsize=(8, 6))
df['Stroke'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Prevalence of Stroke')
plt.axis('equal')
plt.show()

# Plot pie chart for Diabetes
plt.figure(figsize=(8, 6))
df['Diabetes'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Prevalence of Diabetes')
plt.axis('equal')
plt.show()

# Plot pie chart for DiffWalk
plt.figure(figsize=(8, 6))
df['DiffWalk'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Prevalence of Difficulty Walking')
plt.axis('equal')
plt.show()

print("""\nConclusion:

1. High blood pressure is very prevalent, affecting nearly half of the population.

2. High cholesterol is also common, affecting over 40% of individuals.

3. Stroke is relatively rare in this population. This lower prevalence suggests that while strokes are a serious health issue, they are less common compared to other conditions.

4. Type 2 diabetes is a significant condition among those affected.

5. Difficulty Walking condition affects a smaller portion of the population compared to the more common health issues like high blood pressure and high cholesterol. This could be related to various health issues or age-related factors.

""")
```

c. Distribution of Heart Disease (Target Variable):

```python
# Count the occurrences of heart disease cases (1 = Yes, 0 = No)
heart_disease_counts = df['HeartDiseaseorAttack'].value_counts()

# Create a pie chart for heart disease distribution
fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(heart_disease_counts, labels=['No Heart Disease', 'Heart Disease'], autopct='%1.1f%%', colors=['lightgreen', 'salmon'], startangle=90)
ax.set_title('Distribution of Heart Disease Cases')

# Display the pie chart
plt.show()
```

```python
# visualizing the distribution of heart disease cases using a bar chart
heart_disease_counts = df['HeartDiseaseorAttack'].value_counts()

# Create a bar chart for heart disease distribution
fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(heart_disease_counts.index, heart_disease_counts.values, color=['lightgreen', 'salmon'])
ax.set_xticks([0, 1])
ax.set_xticklabels(['No Heart Disease', 'Heart Disease'])
ax.set_title('Distribution of Heart Disease Cases')
ax.set_xlabel('Heart Disease Status')
ax.set_ylabel('Count')

# Display the bar chart
plt.show()
```

# **2.Bivariate analysis**

a. Explore Relationships with Heart Disease:
Compare the distribution of key variables

```python

# Set up the figure and axes for the subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))
fig.suptitle('Comparison of Variables by Heart Disease Status')

# Bar chart for HighBP
sns.countplot(x='HighBP', hue='HeartDiseaseorAttack', data=df, ax=axes[0], palette='Set1')
axes[0].set_title('High Blood Pressure vs Heart Disease')
axes[0].set_xlabel('High Blood Pressure (1 = Yes, 0 = No)')
axes[0].set_ylabel('Count')

# Bar chart for HighChol
sns.countplot(x='HighChol', hue='HeartDiseaseorAttack', data=df, ax=axes[1], palette='Set2')
axes[1].set_title('High Cholesterol vs Heart Disease')
axes[1].set_xlabel('High Cholesterol (1 = Yes, 0 = No)')
axes[1].set_ylabel('Count')

# Box plot for BMI
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=df, ax=axes[2], palette='Set3')
axes[2].set_title('BMI Distribution vs Heart Disease')
axes[2].set_xlabel('Heart Disease (1 = Yes, 0 = No)')
axes[2].set_ylabel('BMI')

# Display the plots
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# b. Visualize Correlations Between Variables:

```python

# List of continuous variables
continuous_vars = ['BMI', 'MentHlth', 'PhysHlth', 'Age']

# Set up the scatter plot matrix (pairplot)
sns.pairplot(df[continuous_vars], diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle('Scatter Plot Matrix of Continuous Variables', y=1.02)
plt.show()

# Calculate the correlation matrix
corr_matrix = df[continuous_vars].corr()

# Plotting the heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# c. Compare Heart Disease Across Demographic Groups:

```python
# Convert demographic groups to categorical if necessary
df['Sex'] = df['Sex'].astype('category')
df['Age'] = df['Age'].astype('category')
df['Education'] = df['Education'].astype('category')

# Create a pivot table to show the prevalence of heart disease across different groups
pivot_table = df.pivot_table(values='HeartDiseaseorAttack',
                               index=['Sex', 'Age', 'Education'],
                               aggfunc='mean')
print("Pivot Table of Heart Disease Prevalence by Gender, Age, and Education:")
print(pivot_table)

# Bar plot for heart disease prevalence by gender
plt.figure(figsize=(10, 5))
# Pass the dataframe to the data argument and specify column names for x and y
sns.barplot(x='Sex', y='HeartDiseaseorAttack', data=df, estimator=lambda x: sum(x)/len(x), ci=None, palette='Set2')
plt.title('Heart Disease Prevalence by Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Prevalence of Heart Disease')
plt.show()

# Bar plot for heart disease prevalence by age group
plt.figure(figsize=(12, 6))
# Pass the dataframe to the data argument and specify column names for x and y
sns.barplot(x='Age', y='HeartDiseaseorAttack', data=df, estimator=lambda x: sum(x)/len(x), ci=None, palette='Set1')
plt.title('Heart Disease Prevalence by Age Group')
plt.xlabel('Age')
plt.ylabel('Prevalence of Heart Disease')
plt.show()

# Bar plot for heart disease prevalence by education level
plt.figure(figsize=(12, 6))
# Pass the dataframe to the data argument and specify column names for x and y
sns.barplot(x='Education', y='HeartDiseaseorAttack', data=df, estimator=lambda x: sum(x)/len(x), ci=None, palette='Set3')
plt.title('Heart Disease Prevalence by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Prevalence of Heart Disease')
plt.show()
```
